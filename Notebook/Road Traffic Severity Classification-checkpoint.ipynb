```python
min = list(range(5,56, 5))
def convert_minutes(x: int):
    for m in min:
        if x % m == x and x > m-5:
            return m
        if x in [56,57,58,59]:
            return 0
        if x in min+[0]:
            return x
```

## Data Preprocessing :

### Handling Missing Values:

```python
df.isna().sum()
```

```python
msno.matrix(df)
```

```python
# dropping columns that can cause imbalance while imputation
df.drop(columns = ['vehicle_defect', 'vehicle_driver_relation', 'casualty_work', 'casualty_fitness'], inplace=True)
```

```python
impute_cols = [x for x in df.isna().sum()[df.isna().sum() != 0].index.tolist()]
for feat in impute_cols:
    mode = df[feat].mode()[0]
    df[feat].fillna(mode, inplace=True)
```

```python
df.isna().sum()
```

### Label Encoding :

Let's use `Label encoding` to encode categorical features,Since the Order does'nt matter here.

```python
#Splitting the Target Variable from the DataFrame :

y = df.accident_severity
del df['accident_severity']
```

```python
object_columns = [col for col in df.columns if df[col].dtypes == 'object']
object_columns
```

```python
for col in object_columns:
    df[col] = LabelEncoder().fit_transform(df[col])

df.head()
```

Splitting the Target Variable from the DataFrame :

### Mutual info Classif() to Check Informative Features :

```python
mutual_info_classifier = mutual_info_classif(df,y,random_state=42)

mutual_info_classifier_df = pd.DataFrame({'Columns':df.columns, 'MI_score':mutual_info_classifier})
mutual_info_classifier_df.sort_values(by='MI_score',ascending=False).head(15)
```

```python
x_train,x_test,y_train,y_test = train_test_split(df,y,test_size=0.3,random_state=42)

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# upsampling using smote

counter = Counter(y_train)

print("=============================")

for k,v in counter.items():
    per = 100*v/len(y_train)
    print(f"Class= {k}, n={v} ({per:.2f}%)")

oversample = SMOTE()
x_train, y_train = oversample.fit_resample(x_train, y_train)

counter = Counter(y_train)

print("=============================")

for k,v in counter.items():
    per = 100*v/len(y_train)
    print(f"Class= {k}, n={v} ({per:.2f}%)")

print("=============================")

print("Upsampled data shape: ", x_train.shape, y_train.shape)
```

```python
y_train = LabelEncoder().fit_transform(y_train)
y_test = LabelEncoder().fit_transform(y_test)
```

### Modelling & Hyper Parameter Tuning :

```python
# Logistic Regression Model:
logreg = LogisticRegression()
logreg.fit(x_train,y_train)

# Predicting on test data using Logistic Regression Model
logreg_prediction = logreg.predict(x_test)

# Evaluating the performance of Logistic Regression Model
Accuracy_lr = accuracy_score(y_test,logreg_prediction)
f1score_logreg = f1_score(y_test,logreg_prediction, average='weighted')
print(f' Accuracy = {Accuracy_lr}, F1_Score = {f1score_logreg}')
```

```python
#Random Forest Base Model:

rfc = RandomForestClassifier()

rfc.fit(x_train,y_train)
prediction = rfc.predict(x_test)

#Evaluation
Accuracy = accuracy_score(y_test,prediction)
f1score = f1_score(y_test,prediction, average='weighted')
print(f' Accuracy = {Accuracy}, F1_Score = {f1score}')
```

```python
#Extra_Tree_Classifer ;

extree = ExtraTreesClassifier()
extree.fit(x_train,y_train)
extree_prediction = extree.predict(x_test)

#Evaluation
extree_Accuracy = accuracy_score(y_test,extree_prediction)
extree_f1score = f1_score(y_test,extree_prediction, average='weighted')
print(f' Accuracy = {extree_Accuracy}, F1_Score = {extree_f1score}')
```

```python
params = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [5, 11, 15],
    'min_samples_split': [2, 3],
    'class_weight': ['balanced', None],
}

# Create a RandomForestClassifier object
rf = RandomForestClassifier()

# Create a GridSearchCV object
gs = GridSearchCV(
    estimator=rf,
    param_grid=params,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=3,
    verbose=3,
)

# Fit the GridSearchCV object to the training data
gs.fit(x_train, y_train)

# Print the best hyperparameters and the best score
print(gs.best_params_)
print(gs.best_score_)
```

```python
# Using the best parameters from GridSearchCV to train the model

rf_tuned = RandomForestClassifier(class_weight=None,
                                max_depth= 15,
                                min_samples_split = 2,
                                n_estimators = 500)
rf_tuned.fit(x_train,y_train)

# Predicting on test data using the best model
tf_tuned_prediction = rf_tuned.predict(x_test)

# Evaluating the performance of the best model
tuned_accuracy = accuracy_score(y_test,tf_tuned_prediction)
tuned_f1score = f1_score(y_test,tf_tuned_prediction, average='weighted')
print(f' Accuracy = {tuned_accuracy}, F1_Score = {tuned_f1score}')
```
